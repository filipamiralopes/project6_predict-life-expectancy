```python
import pandas as pd
import numpy as np
```

```python
life_expectancy = pd.read_csv('./raw_data/life_expectancy_years.csv')
life_expectancy.head()
```

```python
food = pd.read_csv('./raw_data/food_supply_kilocalories_per_person_and_day.csv')
food.head()
```

```python
print(len(food.country.unique()))
```

```python
#subsetting (from 2000 till 2010)
years = [str(i) for i in range(2000,2011)]
years.insert(0,'country')
food_10 = food[[*years]]
food_10.head()
```

```python
print(len(food_10.country.unique()))
print(len(life_expectancy.country.unique()))
```

```python
coutries_life = list(life_expectancy.country.unique())
```

```python
#coutries_life #list_2
countries_income = list(food_10.country.unique()) #list_1
countries_to_drop = np.setdiff1d(coutries_life,countries_income)
countries_to_drop
```

```python
kcal = food_10[food_10['country'].isin(coutries_life)]
kcal.head()
```

```python
coutries_kcal = list(kcal.country.unique())
#print(coutries_gov)
len(coutries_kcal)
```

```python
print(coutries_life==coutries_kcal) #20 countries missing
```

```python
kcal_melt = kcal.melt(id_vars=["country"])
kcal_melt.head()
```

```python
kcal_melt.sort_values(by=['country','variable'], inplace=True)
kcal_melt.head()
```

```python
kcal_day = kcal_melt.rename(columns={'country':'country', 'variable':'year', 'value':'kcal_day'})
kcal_day.reset_index(inplace=True)
kcal_day.drop(columns=['index'], inplace=True)
kcal_day.head()
```
