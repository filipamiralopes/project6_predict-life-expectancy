```python
import pandas as pd
import numpy as np
```

```python
life_expectancy = pd.read_csv('./raw_data/life_expectancy_years.csv')
life_expectancy.head()
```

```python
sugar_person = pd.read_csv('./raw_data/sugar_per_person_g_per_day.csv') 
sugar_person.head()
```

```python
sugar_person.shape
```

```python
#subsetting (from 2000 till 2010)
years = [str(i) for i in range(2000,2011)]
years.insert(0,'country')
sugar_10 = sugar_person[[*years]]
sugar_10.head()
```

```python
print(len(sugar_10.country.unique()))
print(len(life_expectancy.country.unique()))
```

```python
coutries_life = list(life_expectancy.country.unique())
```

```python
#coutries_life #list_2
#countries_income = list(gdpp.country.unique()) #list_1
#countries_to_drop = np.setdiff1d(countries_income,coutries_life)
#countries_to_drop
```

```python
sugar = sugar_10[sugar_10['country'].isin(coutries_life)]
sugar.head()
```

```python
coutries_sugar = list(sugar.country.unique())
print(len(coutries_sugar))
```

```python
print(coutries_life==coutries_sugar) #countries missing in total health spending
```

```python
sugar_melt = sugar.melt(id_vars=["country"])
sugar_melt.head()
```

```python
sugar_melt.sort_values(by=['country','variable'], inplace=True)
sugar_melt.head()
```

```python
sugar_g_day = sugar_melt.rename(columns={'country':'country', 'variable':'year', 'value':'sugar_g_day'})
sugar_g_day.reset_index(inplace=True)
sugar_g_day.drop(columns=['index'], inplace=True)
sugar_g_day.head()
```
