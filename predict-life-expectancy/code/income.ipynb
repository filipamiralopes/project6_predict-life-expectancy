```python
import pandas as pd
import numpy as np
```

```python
life_expectancy = pd.read_csv('./raw_data/life_expectancy_years.csv')
life_expectancy.head()
```

```python
gd_p = pd.read_csv('./raw_data/gdppercapita_us_inflation_adjusted.csv') #US
gd_p.head()
```

```python
gd_p.shape
```

```python
#subsetting (from 2000 till 2010)
years = [str(i) for i in range(2000,2011)]
years.insert(0,'country')
gdpp = gd_p[[*years]]
gdpp.head()
```

```python
print(len(gdpp.country.unique()))
print(len(life_expectancy.country.unique()))
```

```python
coutries_life = list(life_expectancy.country.unique())
```

```python
#coutries_life #list_2
#countries_income = list(gdpp.country.unique()) #list_1
#countries_to_drop = np.setdiff1d(countries_income,coutries_life)
#countries_to_drop
```

```python
gdp = gdpp[gdpp['country'].isin(coutries_life)]
gdp.head()
```

```python
coutries_gdp = list(gdp.country.unique())
```

```python
print(coutries_life==coutries_gdp)
```

```python
gdp_melt = gdp.melt(id_vars=["country"])
gdp_melt.head()
```

```python
gdp_melt.sort_values(by=['country','variable'], inplace=True)
gdp_melt.head(20)
```

```python
income = gdp_melt.rename(columns={'country':'country', 'variable':'year', 'value':'income'})
income.reset_index(inplace=True)
income.drop(columns=['index'], inplace=True)
income.head()
```
