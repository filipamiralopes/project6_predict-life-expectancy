```python
import pandas as pd
import numpy as np
```

```python
life_expectancy = pd.read_csv('./raw_data/life_expectancy_years.csv')
life_expectancy.head()
```

```python
government = pd.read_csv('./raw_data/government_health_spending_per_person_us.csv')
government.head()
```

```python
government.shape
```

```python
#subsetting (from 2000 till 2010)
years = [str(i) for i in range(2000,2011)]
years.insert(0,'country')
gov_10 = government[[*years]]
gov_10.head()
```

```python
print(len(gov_10.country.unique()))
print(len(life_expectancy.country.unique()))
```

```python
coutries_life = list(life_expectancy.country.unique())
```

```python
#coutries_life #list_2
#countries_income = list(gdpp.country.unique()) #list_1
#countries_to_drop = np.setdiff1d(countries_income,coutries_life)
#countries_to_drop
```

```python
gov = gov_10[gov_10['country'].isin(coutries_life)]
gov.head()
```

```python
coutries_gov = list(gov.country.unique())
#print(coutries_gov)
len(coutries_gov)
```

```python
print(coutries_life==coutries_gov) #6 countries missing
```

```python
gov_melt = gov.melt(id_vars=["country"])
gov_melt.head()
```

```python
gov_melt.sort_values(by=['country','variable'], inplace=True)
gov_melt.head()
```

```python
gov_spending = gov_melt.rename(columns={'country':'country', 'variable':'year', 'value':'gov_health_spending'})
gov_spending.reset_index(inplace=True)
gov_spending.drop(columns=['index'], inplace=True)
gov_spending.head()
```
