```python
import pandas as pd
import numpy as np
```

```python
life_expectancy = pd.read_csv('./raw_data/life_expectancy_years.csv')
life_expectancy.head()
```

```python
healthspending = pd.read_csv('./raw_data/total_health_spending_per_person_us.csv') #$US
healthspending.head()
```

```python
healthspending.shape
```

```python
#subsetting (from 2000 till 2010)
years = [str(i) for i in range(2000,2011)]
years.insert(0,'country')
health_spending_10 = healthspending[[*years]]
health_spending_10.head()
```

```python
print(len(health_spending_10.country.unique()))
print(len(life_expectancy.country.unique()))
```

```python
coutries_life = list(life_expectancy.country.unique())
```

```python
#coutries_life #list_2
#countries_income = list(gdpp.country.unique()) #list_1
#countries_to_drop = np.setdiff1d(countries_income,coutries_life)
#countries_to_drop
```

```python
health_spending = health_spending_10[health_spending_10['country'].isin(coutries_life)]
health_spending.head()
```

```python
coutries_health_spending = list(health_spending.country.unique())
print(len(coutries_health_spending))
```

```python
print(coutries_life==coutries_health_spending) #5 countries missing in total health spending
```

```python
health_spending_melt = health_spending.melt(id_vars=["country"])
health_spending_melt.head()
```

```python
health_spending_melt.sort_values(by=['country','variable'], inplace=True)
health_spending_melt.head()
```

```python
spending_health = health_spending_melt.rename(columns={'country':'country', 'variable':'year', 'value':'total_health_spending'})
spending_health.reset_index(inplace=True)
spending_health.drop(columns=['index'], inplace=True)
spending_health.head()
```
